```python
#building DL model 
#DataLoder = loads jpeg,png and xml in format readable to model 
from detecto.core import DataLoader, Model

labels = ['ford','honda','mini_cooper','chevrolet','toyota', 'mercedes','mitsubishi','volkswagen', 'mini cooper']

model = Model(labels)
```
